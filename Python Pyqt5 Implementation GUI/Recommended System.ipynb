### Finding Similaritites 

```python
import pandas as pd


r_cols=['user_id','movie_id','rating']
ratings=pd.read_csv('ratings.data',sep='\t',names=r_cols, usecols=range(3))

m_cols=['movie_id','title']
movies=pd.read_csv('movies.item',sep='|', names=m_cols,encoding='latin-1', usecols=range(2))

ratings=pd.merge(movies,ratings)
```

```python
ratings.head()
```

```python
movie_ratings=ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movie_ratings.head()
```

```python
starwarsRatings=movie_ratings['Star Wars (1977)']
starwarsRatings.head()
```

```python
similar_movies=movie_ratings.corrwith(starwarsRatings)
similar_movies=similar_movies.dropna()

df=pd.DataFrame(similar_movies)

df.head(10)
```

```python
similar_movies.sort_values(ascending=False)
```
